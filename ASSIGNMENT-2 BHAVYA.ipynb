TASK-3 Perform Visualizations

UNIVARIATE ANALYSIS

```python
sns.histplot(df["living area"],bins=30,kde=True)
plt.title("Living Area Analysis",color="red")
plt.xlabel("Living Area",color='green')
plt.ylabel("Count",color="green")
plt.show()
```

```python
sns.boxplot(df['living_area_renov'])
plt.title("Living area renovation")
plt.show()
```

BIVARIATE ANALYSIS

```python
sns.barplot(x='number of bedrooms',y='number of bathrooms',data=df)
plt.title("No of Bedrooms Vs Bathrooms")
plt.show()
```

```python
sns.lineplot(x='number of floors',y='number of views',data=df)
plt.title("No of floors Vs views")
plt.show()
```

MULTIVARIATE ANALYSIS

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True,linewidths=0.8,linecolor="yellow",cmap="coolwarm")
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True,linewidths=0.8,linecolor="black",cmap="jet",cbar=False,center=0.4)
```
